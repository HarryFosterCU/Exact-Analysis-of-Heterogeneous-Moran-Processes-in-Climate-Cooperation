```python
import main
import numpy as np
import sympy as sym
```

```python
def public_goods_fitness_function(state, alpha, r):
    number_of_contributors = np.sum(state)
    total_contribution = number_of_contributors * alpha * r
    try:
        return (np.array([(total_contribution/len(state)) + ((1 - type) * alpha) * 1.001 for type in state]))
    except:
        return (np.array([sym.exp((total_contribution/len(state)) + ((1 -type) * alpha)) * 1.001 for type in state]))
```

```python
state_space = main.get_state_space(N=4, k=2)
sorted_state_space = state_space[np.argsort(state_space.sum(axis=1))]
sorted_state_space
```

```python
alpha = 5
r = 2
transition_matrix = main.generate_transition_matrix(state_space=sorted_state_space, fitness_function=public_goods_fitness_function, alpha=alpha, r=r)
main.get_absorption_probability(transition_matrix=transition_matrix,
                                rounding_places=2)
```
